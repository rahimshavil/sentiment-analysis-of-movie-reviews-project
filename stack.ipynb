```python
stack=[]

print("--- Pushing items---")
stack.append("A")
print("stack:", stack)

stack.append("Bo")
print("stack:", stack)

stack.append("Book C")
print("stack:", stack)

if stack:
    print(f"\nTop item is: {stack[-1]}")


print("\n--- Popping items ---")


reomved_item = stack.pop()
print(f"Removed item: {reomved_item}")
print("stack:", stack)

removed_item  = stack.pop()
print(f"Removed item: {removed_item}")
print("stack:", stack)







```
